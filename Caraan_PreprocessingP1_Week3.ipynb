```python
import pandas as pd
```

```python
# Initialization
sales_data = pd.read_csv("sales.csv")

# Removing the duplicate row values
sales_data = sales_data.drop_duplicates()

# Renaming the column
sales_data.rename(columns={"Price": "Price_K"}, inplace=True)

#since the data type of Price is in object I need it to convert into numeric
sales_data['Price_K'] = pd.to_numeric(sales_data['Price_K'], errors='coerce')

# Dividing the Price_K column by 1000
sales_data["Price_K"] = sales_data["Price_K"] / 1000
```

```python
sales_data
```

#### Problem 1: Select name, city, state and country data of product 2 bought by customers from united states and paid thru Visa or Mastercard. Store the selected data in empty.csv file

```python
filtered_sales_data1 = sales_data[
    (sales_data['Product'] == 'Product2') &
    (sales_data['Country'] == 'United States') &
    (sales_data['Payment_Type'].isin(['Visa', 'Mastercard']))][['Name', 'City', 'State', 'Country']]

filtered_sales_data1
```

```python
filtered_sales_data1.to_csv("empty.csv", index = False)

filtered_sales_data1_result = pd.read_csv("empty.csv")
filtered_sales_data1_result
```

#### Problem 2: Select name, price, city, state and country data of product 2 bought by customers from New York and paid thru Mastercard. Store the selected data in empty1.csv file.

```python
filtered_sales_data2 = sales_data[
    (sales_data['Product'] == 'Product2') &
    (sales_data['City'] == 'New York') &
    (sales_data['Payment_Type'] == 'Mastercard')][['Name','Price','City', 'State', 'Country']]

filtered_sales_data2
```

```python
filtered_sales_data2.to_csv("empty1.csv", index = False)

filtered_sales_data2_result = pd.read_csv("empty1.csv")
filtered_sales_data2_result
```
